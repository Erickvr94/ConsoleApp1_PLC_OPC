```python

# =========================================================
# HELPERS PARA TU DATA
# =========================================================
def iter_tas(ps: str, ps_obj: dict):
    """
    Devuelve (ta_full, ta_short, meta) por tablero real.
    - ta_full  : "PS29_TA001"
    - ta_short : "TA001"
    """
    for k, v in ps_obj.items():
        if k == "NamePis":
            continue
        if not isinstance(v, dict):
            continue

        ta_full = k
        prefix = ps + "_"
        ta_short = ta_full[len(prefix):] if ta_full.startswith(prefix) else ta_full

        yield ta_full, ta_short, v
```

```python
# =========================================================
# LECTURA PLANTILLA + MODELOS
# =========================================================
def load_template_df():
    xls = pd.ExcelFile(PLANTILLA_XLSX)
    sheet = xls.sheet_names[0]
    tpl = pd.read_excel(PLANTILLA_XLSX, sheet_name=sheet, header=0)
    tpl.columns = tpl.columns.astype(str).str.strip()
    return tpl, sheet

def pick_model_rows(template_df: pd.DataFrame):
    """
    Modelos por 'key' (busca substring en Name).
    Además, retorna un modelo especial para las 17 TCP (buscando 'Alarma_Temp_TCP').
    """
    if "Name" not in template_df.columns:
        raise ValueError(f"La plantilla no tiene columna 'Name'. Columnas: {template_df.columns.tolist()}")

    models = {}
    name_series = template_df["Name"].astype(str)

    for d in ALARM_DEFS:
        key = d["key"]
        m = name_series.str.contains(key, na=False)
        models[key] = template_df[m].iloc[0].copy() if m.any() else template_df.iloc[0].copy()

    # Modelo base para las 17 TCP fijas
    m_tcp = name_series.str.contains("Alarma_Temp_TCP", na=False)
    models["_TCP_FIXED_"] = template_df[m_tcp].iloc[0].copy() if m_tcp.any() else template_df.iloc[0].copy()

    return models

def set_if_exists(row: pd.Series, col: str, val):
    if col in row.index:
        row[col] = val
    return row

def make_alarm_row(model_row: pd.Series, alarm_id: int, name: str, text: str, alarm_class: str, trig_tag: str, trig_bit):
    r = model_row.copy()
    r = set_if_exists(r, "ID", int(alarm_id))
    r = set_if_exists(r, "Name", str(name))
    r = set_if_exists(r, "Alarm text [en-US], Alarm text 1", str(text))
    r = set_if_exists(r, "Class", str(alarm_class))
    r = set_if_exists(r, "Trigger tag", str(trig_tag))

    # Trigger bit puede ser vacío o número
    if trig_bit == "" or trig_bit is None:
        r = set_if_exists(r, "Trigger bit", "")
    else:
        r = set_if_exists(r, "Trigger bit", int(trig_bit))

    return r

def compute_circuit_bit(defn: dict, circuit: int) -> int:
    if "trig_bit_map" in defn:
        bit_map = defn["trig_bit_map"]
        if circuit - 1 >= len(bit_map):
            raise ValueError(f"trig_bit_map no cubre el circuito {circuit}. Mapa: {bit_map}")
        return int(bit_map[circuit - 1])

    base = int(defn["trig_bit_base"])
    return base + (circuit - 1)
```

```python
# =========================================================
# GENERADOR
# =========================================================
def generate(id_start: int):
    tpl, _ = load_template_df()
    models = pick_model_rows(tpl)

    out_rows = []
    alarm_id = int(id_start)

    for ps, ps_obj in DATA.items():
        for ta_full, ta_short, meta in iter_tas(ps, ps_obj):

            is_tcp = str(meta.get("IsTCP", "false")).lower() == "true"
            cant   = int(meta.get("CantCirc", 0))

            # TA para Name/Text (normalmente "TA001")
            TA_FOR_NAME = ta_short

            # -------------------------
            # 1) Alarmas generales (ALARM_DEFS)
            # -------------------------
            for d in ALARM_DEFS:
                scope = d["scope"]

                if scope == "TA_CIRCUIT":
                    for c in range(1, cant + 1):
                        name = d["name_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C=c, S="")
                        text = d["text_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C=c, S="")
                        trig_tag = d["trig_tag_tpl"].format(PS=ps, TA=TA_FOR_NAME, TA_FULL=ta_full, C=c, S="")
                        trig_bit = compute_circuit_bit(d, c)

                        out_rows.append(make_alarm_row(
                            models[d["key"]], alarm_id, name, text, d["class_value"], trig_tag, trig_bit
                        ))
                        alarm_id += 1

                elif scope == "TA_SINGLE":
                    name = d["name_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C="", S="")
                    text = d["text_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C="", S="")
                    trig_tag = d["trig_tag_tpl"].format(PS=ps, TA=TA_FOR_NAME, TA_FULL=ta_full, C="", S="")
                    trig_bit = d.get("trig_bit_fixed", "")

                    out_rows.append(make_alarm_row(
                        models[d["key"]], alarm_id, name, text, d["class_value"], trig_tag, trig_bit
                    ))
                    alarm_id += 1

                elif scope == "TA_TCPONLY":
                    if not is_tcp:
                        continue
                    name = d["name_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C="", S="")
                    text = d["text_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C="", S="")
                    trig_tag = d["trig_tag_tpl"].format(PS=ps, TA=TA_FOR_NAME, TA_FULL=ta_full, C="", S="")
                    trig_bit = d.get("trig_bit_fixed", "")

                    out_rows.append(make_alarm_row(
                        models[d["key"]], alarm_id, name, text, d["class_value"], trig_tag, trig_bit
                    ))
                    alarm_id += 1

                elif scope == "TA_STEPS12_TCP":
                    if not is_tcp:
                        continue
                    base = int(d["trig_bit_base"])
                    for s in range(1, 13):
                        name = d["name_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C="", S=s)
                        text = d["text_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=TA_FOR_NAME, C="", S=s)
                        trig_tag = d["trig_tag_tpl"].format(PS=ps, TA=TA_FOR_NAME, TA_FULL=ta_full, C="", S=s)
                        trig_bit = base + (s - 1)

                        out_rows.append(make_alarm_row(
                            models[d["key"]], alarm_id, name, text, d["class_value"], trig_tag, trig_bit
                        ))
                        alarm_id += 1

            # -------------------------
            # 2) Bloque FIJO de 17 alarmas TCP (solo si IsTCP=true)
            # -------------------------
            if is_tcp:
                for a in TCP_FIXED_ALARMS:
                    name = f"{NOMB_PREFIJO}_{a['name_lbl']}_{ps}_{TA_FOR_NAME}"
                    text = f"{a['text_lbl']} {ps}_{TA_FOR_NAME}"
                    trig_tag = f"{ta_full}.{a['alarm']}"  # PS29_TA001.Alarm1 / Alarm2
                    trig_bit = a["bit"]

                    out_rows.append(make_alarm_row(
                        models["_TCP_FIXED_"],
                        alarm_id,
                        name,
                        text,
                        TCP_FIXED_CLASS,
                        trig_tag,
                        trig_bit
                    ))
                    alarm_id += 1

    out_df = pd.DataFrame(out_rows, columns=tpl.columns)

    with pd.ExcelWriter(OUTPUT_XLSX, engine="openpyxl") as w:
        out_df.to_excel(w, sheet_name=OUTPUT_SHEET_NAME, index=False, header=True)

    return OUTPUT_XLSX, len(out_df), (alarm_id - 1)

# =========================================================
# MAIN
# =========================================================
if __name__ == "__main__":
    out, n, last_id = generate(ID_START)
    print("OK ->", out)
    print("Filas generadas:", n)
    print("ID inicio:", ID_START, " | ID final:", last_id)
    print("Hora local:", datetime.now().strftime("%H:%M:%S"))
```
